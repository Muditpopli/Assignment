Information Gain is a measure used in decision trees to determine the effectiveness of splitting a node based on a particular attribute. It quantifies the reduction in entropy (or impurity) achieved by partitioning the data according to that attribute.

Information gain acts as a heuristic to guide the decision tree learning algorithm in selecting the most informative features to partition the data at each step, aiming to create a tree that effectively classifies or predicts the target variable.

Answer to Q.No 7:

The mathematical explanation behind Decision Trees lies in recursively partitioning the feature space into non-overlapping regions. At each internal node, the algorithm selects an optimal split point on a chosen feature to maximize the separation of data points belonging to different classes (for classification) or to reduce the variance of the target variable (for regression). This optimal split is determined by evaluating impurity measures like Entropy or Gini Impurity (for classification) or variance reduction (for regression) across potential splits and selecting the one that yields the greatest improvement. The process continues down the tree, creating a hierarchical structure where each leaf node represents a region in the feature space associated with a specific prediction. Ultimately, the tree approximates a piecewise constant function that maps input features to output predictions.

Answer to Q.No 9:

Decision trees are predictive models employing a hierarchical, tree-like structure to classify or regress data by recursively splitting it based on feature values. Impurity measures like Gini Impurity and Entropy quantify the homogeneity of data subsets, guiding the selection of optimal splits that maximize information gain. While Gini Impurity calculates the probability of misclassification, Entropy measures data disorder. The mathematical foundation of decision trees lies in this recursive partitioning of the feature space into distinct regions, each associated with a specific prediction. To prevent overfitting, two primary strategies exist: pre-pruning, which halts tree growth early based on predefined criteria, and post-pruning, which involves growing a full tree and then pruning branches that do not enhance generalization on unseen data. Ultimately, decision trees offer an interpretable and versatile approach to machine learning.

Answer to Q.No 10:

Pre-pruning and post-pruning are both techniques to combat overfitting in decision trees, but they differ in their approach: pre-pruning stops the tree's growth during construction by setting criteria that prevent further splitting of nodes, while post-pruning allows the tree to grow fully and then removes branches or subtrees that do not improve performance on a validation set, effectively simplifying the model after it has been fully formed.

The goal is to find the categorical split that best separates the data according to the chosen impurity measure (for classification) or variance reduction (for regression).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)

# Train the classifier on the training data
path = clf.cost_complexity_pruning_path(X_train, y_train)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

# We will store the classifiers trained with different ccp_alphas
clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=42, ccp_alpha=ccp_alpha)
    clf.fit(X_train, y_train)
    clfs.append(clf)

# Remove the last element as it's the trivial tree with only the root node
clfs = clfs[:-1]
ccp_alphas = ccp_alphas[:-1]

# Evaluate the accuracy of each pruned tree on the test set
train_scores = [clf.score(X_train, y_train) for clf in clfs]
test_scores = [clf.score(X_test, y_test) for clf in clfs]

# Plot the accuracy vs. ccp_alpha
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("Accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker='o', label="test", drawstyle="steps-post")
ax.legend()
plt.show()

# Find the index of the best performing tree on the test set
best_alpha_index = np.argmax(test_scores)
best_ccp_alpha = ccp_alphas[best_alpha_index]
best_clf = clfs[best_alpha_index]

print(f"\nBest ccp_alpha: {best_ccp_alpha:.4f}")
print(f"Accuracy of the best pruned tree on the test set: {best_clf.score(X_test, y_test):.4f}")

# Optional: Visualize the best pruned tree
plt.figure(figsize=(12, 8))
plot_tree(best_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title(f"Best Pruned Decision Tree (ccp_alpha={best_ccp_alpha:.4f})")
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid to search
param_grid = {
    'max_depth': [None, 3, 5, 7, 10],
    'min_samples_split': [2, 5, 10, 15, 20]
}

# Create a Decision Tree Classifier
dtc = DecisionTreeClassifier(random_state=42)

# Set up GridSearchCV
grid_search = GridSearchCV(estimator=dtc,
                           param_grid=param_grid,
                           cv=5,  # Number of cross-validation folds
                           scoring='accuracy',
                           verbose=2,
                           n_jobs=-1)  # Use all available cores

# Perform the grid search on the training data
grid_search.fit(X_train, y_train)

# Print the best parameters found
print("Best Parameters found by GridSearchCV:")
print(grid_search.best_params_)

# Get the best estimator (the model with the best parameters)
best_dtc = grid_search.best_estimator_

# Evaluate the best model on the test set
y_pred_best = best_dtc.predict(X_test)
accuracy_best = accuracy_score(y_test, y_pred_best)
print(f"Accuracy of the best model on the test set: {accuracy_best:.4f}")

# Evaluate the default model for comparison
default_dtc = DecisionTreeClassifier(random_state=42)
default_dtc.fit(X_train, y_train)
y_pred_default = default_dtc.predict(X_test)
accuracy_default = accuracy_score(y_test, y_pred_default)
print(f"Accuracy of the default model on the test set: {accuracy_default:.4f}")

# Compare the accuracies
print("\nComparison:")
if accuracy_best > accuracy_default:
    print("The model tuned with GridSearchCV performed better than the default model.")
elif accuracy_default > accuracy_best:
    print("The default model performed better than the model tuned with GridSearchCV.")
else:
    print("Both models have the same accuracy.")
```
