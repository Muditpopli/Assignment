Answer to Q. No.6:

Concepts of Generators in Python:

Generators are special functions that create a sequence of values one at a time, as needed.
They use yield keyword instead of return to return value.
For example:


```python
def countdown(start):
  while start >= 0:
    yield start
    start -= 1

for number in countdown(5):
  print(number)
```

```python
def sum_even_numbers(numbers):
  even_numbers = [num for num in numbers if num % 2 == 0]  # Using list comprehension for efficiency
  return sum(even_numbers)

numbers = [1, 2, 3, 4, 5]
result = sum_even_numbers(numbers)
print(result)
```

```python
#Answer to Q.No 4

def is_prime(num):
  if num <= 1:
    return False
  for i in range(2, int(num**0.5) + 1):
    if num % i == 0:
      return False
  return True

print(is_prime(17))
```

```python
#Answer to Q no 5:

def fibonacci(number):
  a, b = 0, 1
  for _ in range(number):
    yield a
    a, b = b, a + b

for num in fibonacci(10):
  print(num)
```

```python
#Answer to Q.No 6:

def powers_of_2(exponent):
  for i in range(exponent + 1):
    yield 2**i


#example
exponent = 5
for power in powers_of_2(exponent):
  print(power)
```

```python
#Answer to Q.No 7:

def read_file_lines(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File '{filename}' not found. Please ensure the file exists and the path is correct.")
        return []
filename = "your_file.txt"  # or "/path/to/your_file.txt"
for line in read_file_lines(filename):
  print(line)
```

```python
#Answer to Q.No 9:

def celsius_to_fahrenheit(celsius):
  return (celsius * 9/5) + 32

temperatures_celsius = [25, 30, 35]
temperatures_fahrenheit = map(celsius_to_fahrenheit, temperatures_celsius)
print(list(temperatures_fahrenheit))  # Output: [77.0, 86.0, 95.0]
```

```python
#Answer to Q.No. 10:

def remove_vowels(string):
  vowels = "aeiouAEIOU"
  return "".join(char for char in string if char not in vowels)

string = "hello, world!"
filtered_string = filter(lambda char: char not in "aeiouAEIOU", string)
print("".join(filtered_string))
```

```python
#Answer to Q.No 11:

def process_orders(orders):
  return [(order[0], order[3] * order[2] + 10 if order[3] * order[2] < 100 else order[3] * order[2]) for order in orders]

orders = [
    (34567, "Learning Python, Mark Lutz", 4, 56.99),
    (77766, "Head First Python, Paul Barry", 3, 36.80),
    (88772, "Programming in Python, Mark Klein", 5, 24.95)
]

processed_orders = process_orders(orders)
print(processed_orders)
```
